```python
import yfinance as yf
import pandas as pd

# Extraindo os dados da Tesla
tesla = yf.download('TSLA', start='2023-01-01', end='2023-12-31')

# Redefinindo o índice para a coluna 'Date'
tesla = tesla.reset_index()

# Exibindo as primeiras 5 linhas
tesla.head()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Assumindo que você já tem os dados da Tesla em um DataFrame chamado 'tesla_data'
# e que o índice é a data

def make_graph(data, title):
  """
  Cria um gráfico de linha para os dados fornecidos.

  Args:
    data: Um DataFrame do Pandas contendo os dados a serem plotados.
    title: O título do gráfico.
  """
  plt.figure(figsize=(12, 6))
  plt.plot(data['Close'])
  plt.title(title)
  plt.xlabel('Data')
  plt.ylabel('Preço de Fechamento')
  plt.grid(True)
  plt.show()

# Chamando a função para criar o gráfico
make_graph(tesla, "Preço de Fechamento das Ações da Tesla")
```
